### d) Most popular product category from customer orders

```python
orders=[
    {"product":"shirt","category":"clothing"},
    {"product":"laptop","category":"electronics"},
{"product":"pants","category":"clothing"},
{"product":"phone","category":"electronics"},
{"product":"jacket","category":"clothing"}
]
category_count={}
for order in orders:
    cat=order["category"]
    category_count[cat]=category_count.get(cat,0)+1
    most_popular=max(category_count,key=category_count.get)
    print("Most popular category:",most_popular)
```

 ### e) student grades dictionary and average:

```python
grades={
    "Ali":85,
    "Sara":92,
"Ahmad":78,
"Zara":90
}
average=sum(grades.values())/len(grades)
print("Grades:",grades)
print("Average grades:",average)        #86.25
```

```python
# --- Student data ---
gradebook = {
    "Ali":   {"Math": 85, "English": 90, "Science": 78},
    "Sara":  {"Math": 92, "English": 88, "Science": 95},
    "Ahmed": {"Math": 70, "English": 75, "Science": 80},
    "Zara":  {"Math": 95, "English": 92, "Science": 89}
}

# --- 1. Calculate average grade for each student ---
def student_averages():
    averages = {}
    for student, subjects in gradebook.items():
        avg = sum(subjects.values()) / len(subjects)
        averages[student] = round(avg, 2)
    return averages

# --- 2. Find highest and lowest in a specific subject ---
def highest_lowest(subject):
    scores = {name: grades[subject] for name, grades in gradebook.items()}
    highest = max(scores, key=scores.get)
    lowest  = min(scores, key=scores.get)
    print(f"Subject: {subject}")
    print(f"  Highest: {highest} ({scores[highest]})")
    print(f"  Lowest:  {lowest}  ({scores[lowest]})")

# --- 3. Sort students by average grade ---
def sort_by_average():
    averages = student_averages()
    sorted_students = sorted(averages.items(), key=lambda x: x[1], reverse=True)
    print("\nStudents ranked by average:")
    for rank, (name, avg) in enumerate(sorted_students, 1):
        print(f"  {rank}. {name}: {avg}")

# -------- Run --------
print("Averages:", student_averages())
highest_lowest("Math")
sort_by_average()




```

```python
# --- Product catalog ---
store = {
    "shirt":  {"price": 800,  "quantity": 10},
    "pants":  {"price": 1200, "quantity": 5},
    "shoes":  {"price": 2500, "quantity": 8},
    "cap":    {"price": 400,  "quantity": 15}
}

# Shopping cart: {product_name: quantity}
cart = {}

# --- 1. Add product to cart ---
def add_to_cart(product, qty=1):
    if product not in store:
        print(f"'{product}' does not exist in store.")
        return
    if store[product]["quantity"] < qty:
        print(f"Not enough stock for '{product}'.")
        return
    cart[product] = cart.get(product, 0) + qty
    print(f"Added {qty} x {product} to cart.")

# --- 2. Remove product from cart ---
def remove_from_cart(product):
    if product in cart:
        del cart[product]
        print(f"'{product}' removed from cart.")
    else:
        print(f"'{product}' not in cart.")

# --- 3. Calculate total cost ---
def total_cost():
    total = 0
    print("\n--- Cart Summary ---")
    for product, qty in cart.items():
        price = store[product]["price"]
        subtotal = price * qty
        print(f"  {product:10} x{qty}  Rs.{price} = Rs.{subtotal}")
        total += subtotal
    print(f"  {'Raw Total':14}        Rs.{total}")
    return total

# --- 4. Apply discount and tax ---
def final_bill(discount=10, tax=5):
    total = total_cost()

    discount_amt = total * discount / 100
    after_discount = total - discount_amt

    tax_amt = after_discount * tax / 100
    final = after_discount + tax_amt

    print(f"\n  Discount ({discount}%):  - Rs.{discount_amt:.2f}")
    print(f"  Tax     ({tax}%):    + Rs.{tax_amt:.2f}")
    print(f"  Final Bill:          Rs.{final:.2f}")

# -------- Test it --------
add_to_cart("shirt", 2)
add_to_cart("shoes", 1)
add_to_cart("cap",   3)
remove_from_cart("cap")

final_bill(discount=10, tax=5)
```
