### Greeting with time:

```python
from datetime import datetime

def greet_with_time(name):
    now = datetime.now().strftime("%H:%M:%S")
    print(f"Hello, {name}! The current time is {now}.")

greet_with_time("Alice")
```

### Question 3

### Calculate discount:

```python
def calculate_discount(price, discount=10):
    discounted = price - (price * discount / 100)
    return discounted

print(calculate_discount(100))       
print(calculate_discount(100, 25))    
```

### Question 5
### Find maximum:

```python
def find_max(*args):
    if not args:
        return None
    return max(args)

print(find_max(3, 7, 1, 9, 4))
print(find_max(100, 42))
```

### Question 6
### Factorial (recursive):

```python
def factorial(n):
    if n == 0 or n == 1:   # base case
        return 1
    return n * factorial(n - 1)  # recursive case

print(factorial(5))   # 5! = 120
print(factorial(0))   # 0! = 1

```

### Question 7
### Function as argument:

```python
def apply_function(func, numbers):
    return [func(n) for n in numbers]

def square(x):
    return x ** 2

result = apply_function(square, [1, 2, 3, 4, 5])
print(result)
```

### Question 8
### Lambda Function:

```python
square = lambda x: x ** 2

print(square(5))     # 25
print(square(9))     # 81

# Also useful inline:
nums = [3, 1, 4, 1, 5]
print(sorted(nums, key=lambda x: -x))
```

### Question 9
### Apply operation:

```python
def apply_operation(func, numbers):
    return list(map(func, numbers))

# double each number
result = apply_operation(lambda x: x * 2, [1, 2, 3, 4, 5])
print(result)

# cube each number
cubed = apply_operation(lambda x: x**3, [2, 3, 4])
print(cubed)
```

### Question 10
### Execution Time Decorator:

```python
import time

def timer(func):
    def wrapper(*args, **kwargs):
        start = time.time()
        result = func(*args, **kwargs)
        end = time.time()
        print(f"{func.__name__} took {end - start:.4f}s")
        return result
    return wrapper

@timer
def slow_sum(n):
    time.sleep(0.1)
    return sum(range(n))

print(slow_sum(1000))
```
