### Question 3

### Sum of N numbers:


```python
n = int(input("Enter a positive integer: "))
total = 0
i = 1
while i <= n:
    total = total + i
    i = i + 1
print("Sum =", total)
```

### Question 8

###  Running Total Until User Stops:

```python
total = 0
num = int(input("Enter a number (0 to stop): "))
while num != 0:
    total = total + num
    num = int(input("Enter a number (0 to stop): "))
print("Final total:", total)
```
